```python
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd

data_top_5 = pd.read_csv(r"extra/data/best_5_cities.csv")
data_cities = pd.read_csv(r"extra/data/city_coords.csv")
data_top_5 = data_top_5.merge(data_cities, on='city')

# Top 5 Cities Map
fig = go.Figure()
fig.add_trace(go.Scattermapbox(
    lat=data_top_5['lat'],
    lon=data_top_5['lon'],
    marker=dict(size=20),
    text=data_top_5['city'],
    hoverinfo='text'
))

fig.update_layout(
    mapbox_style="open-street-map",
    margin=dict(l=0, r=0, t=20, b=0),
    mapbox=dict(
        zoom=4.5,
        center=dict(lat=46, lon=2)
        ),
    width=600,
    height=600
    )

fig.show()
```

```python
# Top 20 Hotels in Best City Map
clean_data = pd.read_csv(r"extra/data/clean_data.csv")
best_city = data_top_5.iloc[0,0]

top_hotels = clean_data[clean_data['city'] == best_city]

import requests
url = 'https://nominatim.openstreetmap.org/search'
headers = {'User-Agent': 'Chrome/123.0.0.0'}

for i, address in enumerate(top_hotels['address']):
    payload = {'q' : address, 'format' : 'json'}
    response = requests.get(url, params=payload, headers=headers).json()

    if len(response) > 0:
        lat = response[0]['lat']
        lon = response[0]['lon']
    else:
        lat = None
        lon = None
    
    top_hotels.loc[i, 'lat'] = lat
    top_hotels.loc[i, 'lon'] = lon

fig = go.Figure()
fig.add_trace(go.Scattermapbox(
    lat=top_hotels['lat'],
    lon=top_hotels['lon'],
    marker=dict(size=20),
    text=top_hotels['hotel_name'],
    hoverinfo='text'
))

fig.update_layout(
    mapbox_style="open-street-map",
    margin=dict(l=0, r=0, t=20, b=0),
    mapbox=dict(
        zoom=14,
        center=dict(lat=42.526, lon=3.082)
        ),
    width=600,
    height=600
    )

fig.show()
```
